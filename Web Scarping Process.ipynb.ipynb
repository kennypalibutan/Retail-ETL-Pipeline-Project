### A. EXTRACT

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()
# Pembuatan kolom kosong yang digunakan untuk menambahkan dari data yang sudah dilooping
nama_produk = []
harga_produk = []
rating_produk = []
foto_produk = []

# Proses pengambilan data dari halaman 1-4 dari website yang dipilih
for hal in range (1,5): # Loop untuk pengambilan data dihalaman yang ditentukan
    url = f'https://jdsports.id/search/shoes?page={hal}'
    driver.get(url)
    time.sleep(5) # Untuk memberikan jeda waktu 5 detik membuka website sebelum dicapture

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    boxes = soup.find_all('div',{'class':'relative cursor-pointer card-product h-full bg-white'})

# Query untuk melakukan looping dan mengakses satu per satu data dan memasukan ke dalam lists kosong yang sudah dibuat
    for box in boxes:
        nama_ecer = box.find('div',{'class':'text-sm w-[85%] ellipsis-2 md:text-xs sm:text-[10px] hover:underline'})
        try:
            nama_produk.append(nama_ecer.get_text())
        except:
            nama_produk.append(None)

        harga_ecer = box.find('span',{'class':'font-bold'})
        try:
            harga_produk.append(harga_ecer.get_text())
        except:
            harga_produk.append(None)

        rating_ecer = box.find('p',{'class':'text-[12px] sm:text-[10px] font-normal tracking-[0.1rem]'})
        try:
            rating_produk.append(rating_ecer.get_text())
        except:
            rating_produk.append(None)

        foto_ecer = box.find('img',{'class':'absolute inset-0 z-0 object-cover object-center w-full h-full opacity-0 group-hover:opacity-100 transition-opacity duration-500'})
        try:
            foto_produk.append(foto_ecer['src'])
        except:
            foto_produk.append(None)
      
# Membuat dataframe untuk diisi dengan lists yang sudah mempunyai data dari looping
df = pd.DataFrame({
'Produk' : nama_produk,
'Harga' : harga_produk,
'Rating' : rating_produk,
'Foto' : foto_produk
})

# Query untuk menyimpan data yang sudah di scraping dari website ke dalam format csv
df.to_csv('CODA_P0M1_Kenny-Palibutan_KOTOR.csv')
```

### B. TRANSFORM

### Query untuk menghapus data yang mempunyai value tidak sesuai

```python
df[~(df['Harga']=='JD Exclusive')].head(50)
df = df[~(df['Harga']=='JD Exclusive')].head(50)
```

```python
df
```

### Query untuk merubah format penulisan kolom harga menjadi hanya angka dengan tipe data integer

```python
df['Harga'].str.replace('Rp','')
df['Harga'] = df['Harga'].str.replace('Rp','')
df['Harga'].str.replace('.','')
df['Harga'] = df['Harga'].str.replace('.','')
df['Harga'].astype(int)
df['Harga'] = df['Harga'].astype(int)

df
```

### Query untuk merubah format penulisan kolom Rating menjadi hanya angka dengan tipe data integer

```python
df['Rating'].str.split('/').str[0]
df['Rating'] = df['Rating'].str.split('/').str[0]
df['Rating'].str.replace('(','')
df['Rating'] = df['Rating'].str.replace('(','')
df['Rating'] = df['Rating'].fillna(0).astype(int)
```
